```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("vine_table.csv")
```

```python
df_rename = df.rename(columns={'RDIJS7QYB6XNR':'review_id',
                         '5': "star_rating",
                         "0": "helpful_votes",
                  "0.1":"total_votes",
                  "N":"vine",
                  "Y":"verified_purchase"})
```

```python
# Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20
df_greater_20 = df_rename[df_rename["total_votes"]>=20]
df_greater_20.head()
```

```python
df_greater_20["helpful_div_total"] = df_greater_20["helpful_votes"] / df_greater_20["total_votes"]
```

```python
# create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
df_greater_50 = df_greater_20[df_greater_20["helpful_div_total"]>=0.5]
df_greater_50.head()
```

```python
# filter df_greater_50 where vine is yes
yes_vine = df_greater_50[df_greater_50["vine"] == 'Y']
print(yes_vine)
```

```python
# filter df_greater_50 where vine is no
no_vine = df_greater_50[df_greater_50["vine"] == 'N']
print(no_vine)
```

```python
# get the total number of 5 star reviews
five_star_ratings = df_greater_20[df_greater_20.star_rating == 5].count()['review_id']

# count number of paid 5 stars reviews
five_star_ratings_paid = df_greater_20[df_greater_20.verified_purchase == "Y"][df_greater_20.star_rating == 5].count()['review_id']
# get percentage
percent_five_stars_ratings_paid = five_star_ratings_paid / five_star_ratings 
print(percent_five_stars_ratings_paid)

# count number of unpaid 5 stars reviews
five_star_ratings_unpaid = df_greater_20[df_greater_20.verified_purchase == "N"][df_greater_20.star_rating == 5].count()['review_id']
# get percentage
percent_five_stars_ratings_unpaid = five_star_ratings_unpaid / five_star_ratings 
print(percent_five_stars_ratings_unpaid)
```
